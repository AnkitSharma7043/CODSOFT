```python
data['Year'] = pd.to_numeric(data['Year'].astype(str).str.extract(r'(\d+)', expand=False), errors='coerce')
```

```python
data['Year'].info()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=data['Year'].value_counts().sort_index())
tick_position=range(min(data['Year']),max(data['Year'])+1,5)
plt.title("Annual Movie Release Counts Over Time")
plt.xticks(tick_position)
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Year',data=data)
plt.title('Boxplot of year')
plt.xlabel('Year')
plt.ylabel('Count')

plt.tight_layout() # adjust layout to prevent clipping of labels
plt.show()
```

```python
data['Year'].median()
```

```python
sns.histplot(data['Year'],bins=50,kde=True,color='red')
plt.title('Histogram of year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(rotation=90) # rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
data['Duration'] = pd.to_numeric(data['Duration'].astype(str).str.extract(r'(\d+)', expand=False), errors='coerce')
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=data.groupby('Year')['Duration'].mean().reset_index(),x='Year',y='Duration')
tick_positions=range(min(data['Year']),max(data['Year']) + 1,5)
plt.title('Average Movie Duration Over the Years')
plt.xticks(tick_positions)
plt.xlabel('Years')
plt.ylabel('Average Duration( in minutes)')
plt.show()
```

```python
sns.boxplot(data=data,y='Duration')
plt.title('Box plot of Average Movie Durations')
plt.ylabel('Average Duration( in minutes)')
plt.show()
```

```python
Q1 = data['Rating'].quantile(0.25)
Q3 = data['Rating'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['Rating'] >= lower_bound) & (data['Rating'] <= upper_bound)]
data.head(5)
```

```python
sns.kdeplot(data['Duration']) # it is normal distribution
```

```python
genre=data['Genre']
genre
```

```python
genres=data['Genre'].str.split(',',expand=True)# expand genre columns
genres
```

```python
stack_data=genres.stack()
stack_data.dropna()
value_count=stack_data.value_counts() # create a data from with frequency corresponding genre
print(value_count)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=value_count.index,y=value_count.values)
plt.title('Counts of Movie Genres')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
data['Votes']=data['Votes'].str.replace(',','').astype(int)
```

```python
actors=pd.concat([data['Actor 1'],data['Actor 2'],data['Actor 3']]).dropna().value_counts()
actors
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=actors.head(20).index,hue=actors.head(20).values,ec='black',palette='rainbow')
plt.xlabel('Actors')
plt.ylabel('Frequency of Movies')
plt.title(' Top 20 Actors by Frequency of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
df_copy = data.copy()
df_copy['Actor'] = df_copy['Actor 1'] + ',' + df_copy['Actor 2'] + ',' + df_copy['Actor 3']
df_copy['Directors'] = df_copy['Director'].astype('category').cat.codes
df_copy['Genres'] = df_copy['Genre'].astype('category').cat.codes
df_copy['Actors'] = df_copy['Actor'].astype('category').cat.codes

df_copy.head(10)
```

```python
sns.histplot(data=df_copy,x='Rating',bins=20,kde=True)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of movie Ratings')
plt.show()
```

```python
sns.boxplot(data=df_copy,y='Rating')
plt.title('Box Plot rating of Movies')
plt.ylabel('frequency')
plt.show()
          
```
